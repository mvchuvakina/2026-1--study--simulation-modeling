# Модель Daisyworld - Анимация динамики

**Цель работы:** Создать анимацию, показывающую эволюцию модели Daisyworld
во времени, чтобы наглядно продемонстрировать процесс саморегуляции.

## 1. Зачем нужна анимация?

Агентные модели особенно наглядны, когда мы можем наблюдать их динамику
в реальном времени. Анимация позволяет увидеть:
* Как маргаритки распространяются по поверхности
* Как изменяется температура в разных областях
* Как формируются паттерны распределения маргариток
* Как система реагирует на изменения солнечной активности

## 2. Инициализация проекта и загрузка пакетов

```julia
using DrWatson
@quickactivate "project"

using Agents
using CairoMakie

include(srcdir("daisyworld.jl"))
```

## 3. Создание модели и настройка визуализации

Создаем модель с параметрами по умолчанию

```julia
model = daisyworld(seed=165)
```

Функция для определения цвета маргаритки

```julia
daisycolor(a::Daisy) = a.breed == :white ? :white : :black
```

Настройки визуализации (аналогичные базовой визуализации)

```julia
plotkwargs = (
    agent_color = daisycolor,
    agent_size = 20,
    agent_marker = '✩',
    heatarray = :temperature,
    heatkwargs = (colorrange = (-20, 60),)
)
```

## 4. Создание анимации

```julia
println("="^60)
println("МОДЕЛЬ DAISYWORLD - АНИМАЦИЯ")
println("="^60)
println("\n🎬 Начало создания анимации...")
println("   - Количество кадров: 60")
println("   - Название: 'Daisy World'")
println("   - Формат: MP4")
```

Функция abmvideo создает видео, последовательно применяя шаги модели
и сохраняя каждый кадр.

```julia
abmvideo(
    plotsdir("daisyworld_simulation.mp4"),  # путь для сохранения
    model;                                   # модель
    title = "Daisy World",                    # заголовок видео
    frames = 60,                               # количество кадров
    plotkwargs...                              # параметры визуализации
)

println("\n✅ Анимация успешно создана!")
println("   📁 Путь: ", plotsdir("daisyworld_simulation.mp4"))
println("\n📌 Рекомендации по просмотру:")
println("   - Обратите внимание на изменение цвета фона (температуры)")
println("   - Наблюдайте за распространением черных и белых маргариток")
println("   - Видите ли вы, как система приходит к равновесию?")
println("\n" * "="^60)
```
