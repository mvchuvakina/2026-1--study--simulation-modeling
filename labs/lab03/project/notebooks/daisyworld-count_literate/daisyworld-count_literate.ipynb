# Модель Daisyworld - Анализ численности маргариток

**Цель работы:** Исследовать, как изменяется численность черных и белых
маргариток с течением времени, и выявить закономерности саморегуляции.

## 1. Теоретическое обоснование

В модели Daisyworld численность маргариток не постоянна. Она меняется
под влиянием температуры, которая, в свою очередь, зависит от соотношения
черных и белых маргариток. Это создает обратную связь:

* Много черных маргариток → повышение температуры → белым маргариткам
  становится комфортнее, они начинают размножаться
* Много белых маргариток → понижение температуры → черным маргариткам
  становится комфортнее

В результате система стремится к равновесию, где численности обоих видов
колеблются вокруг некоторых значений, поддерживая температуру в пригодном
для жизни диапазоне.

## 2. Инициализация проекта и загрузка пакетов

```julia
using DrWatson
@quickactivate "project"

using Agents
using CairoMakie
using DataFrames
using Statistics

include(srcdir("daisyworld.jl"))
```

## 3. Определение функций для сбора данных

Функции-предикаты для фильтрации агентов по виду
Они возвращают true, если агент принадлежит соответствующему виду

```julia
black(a) = a.breed == :black
white(a) = a.breed == :white
```

Спецификация данных для сбора:
- Для черных маргариток собираем количество (count)
- Для белых маргариток собираем количество (count)

```julia
adata = [(black, count), (white, count)]
```

## 4. Запуск симуляции и сбор данных

```julia
println("="^60)
println("МОДЕЛЬ DAISYWORLD - АНАЛИЗ ЧИСЛЕННОСТИ")
println("="^60)
```

Создаем модель с начальной солнечной активностью 1.0

```julia
model = daisyworld(solar_luminosity = 1.0, seed=165)
println("\n🔄 Создана модель Daisyworld")
println("   - Начальная солнечная активность: 1.0")
println("   - Начальная доля черных: 20%")
println("   - Начальная доля белых: 20%")
```

Запускаем модель на 1000 шагов и собираем данные

```julia
println("\n⏳ Запуск симуляции на 1000 шагов...")
println("   (это может занять некоторое время)")
agent_df, model_df = run!(model, 1000; adata = adata)

println("✅ Симуляция завершена!")
println("   - Собрано строк данных: ", nrow(agent_df))
```

## 5. Визуализация результатов

```julia
println("\n📊 Создание графика численности...")
```

Создаем фигуру для графика

```julia
figure = Figure(size = (600, 400))
ax = Axis(figure[1, 1],
    xlabel = "Время, шаги модели",
    ylabel = "Количество маргариток",
    title = "Динамика численности маргариток в модели Daisyworld"
)
```

Строим линии для черных и белых маргариток

```julia
blackl = lines!(ax, agent_df[!, :time], agent_df[!, :count_black],
    color = :black,
    linewidth = 2,
    label = "Черные маргаритки"
)

whitel = lines!(ax, agent_df[!, :time], agent_df[!, :count_white],
    color = :orange,
    linewidth = 2,
    label = "Белые маргаритки"
)
```

Добавляем легенду

```julia
Legend(figure[1, 2], [blackl, whitel], ["Черные", "Белые"])
```

Сохраняем график

```julia
save(plotsdir("daisy_count.png"), figure)
println("✅ График сохранён: ", plotsdir("daisy_count.png"))
```

## 6. Анализ полученных данных

```julia
println("\n📈 Анализ результатов:")
println("   - Средняя численность черных: ", round(mean(agent_df[!, :count_black]), digits=1))
println("   - Средняя численность белых: ", round(mean(agent_df[!, :count_white]), digits=1))
println("   - Максимальная численность черных: ", maximum(agent_df[!, :count_black]))
println("   - Максимальная численность белых: ", maximum(agent_df[!, :count_white]))
println("   - Минимальная численность черных: ", minimum(agent_df[!, :count_black]))
println("   - Минимальная численность белых: ", minimum(agent_df[!, :count_white]))

println("\n📌 Наблюдения:")
println("   - Видны ли колебания численности?")
println("   - Какой вид преобладает?")
println("   - Есть ли корреляция между численностью черных и белых?")
println("\n" * "="^60)
```
