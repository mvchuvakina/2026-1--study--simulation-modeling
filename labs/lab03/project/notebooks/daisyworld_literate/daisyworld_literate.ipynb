## 1. Теоретическое введение

### 1.1 Гипотеза Геи

Гипотеза Геи, предложенная Джеймсом Лавлоком, рассматривает планету как единую,
саморегулирующуюся систему, включающую как живые, так и неживые части.

### 1.2 Модель Daisyworld

В мире маргариток произрастают чёрные и белые маргаритки. Их альбедо различается:
* Чёрные маргаритки поглощают свет и тепло, нагревая окружающую среду.
* Белые маргаритки отражают свет, охлаждая окружающую среду.

Маргаритки могут размножаться только в определённом температурном диапазоне.
Поверхность Геи нагревается солнцем, но растущие на ней маргаритки поглощают
или отражают свет, изменяя локальную температуру.

Взаимодействие живых и неживых аспектов этого мира находит равновесие в широком
диапазоне параметров, демонстрируя саморегуляцию системы.

## 3. Настройка визуализации

Создаем модель с параметрами по умолчанию

```julia
model = daisyworld()
```

Функция для определения цвета маргаритки при визуализации

```julia
daisycolor(a::Daisy) = a.breed == :white ? :white : :black
```

Настройки для визуализации:
- agent_color: цвет агента в зависимости от вида
- agent_size: размер символа агента
- agent_marker: символ для отображения агента
- heatarray: показываем температуру как тепловую карту
- heatkwargs: диапазон цветов для температурной шкалы

```julia
plotkwargs = (
    agent_color = daisycolor,
    agent_size = 20,
    agent_marker = '✩',
    heatarray = :temperature,
    heatkwargs = (colorrange = (-20, 60),)
)
```

## 4. Визуализация на разных шагах модели

### Шаг 1: Начальное состояние

```julia
println("Шаг 1: Начальное состояние")
plt1, _ = abmplot(model; plotkwargs...)
save(plotsdir("daisy_step001.png"), plt1)
println("✅ График сохранён: ", plotsdir("daisy_step001.png"))
```

### Шаг 2: После 5 итераций

```julia
step!(model, 5)
println("\nШаг 5: После 5 итераций")
plt2, _ = abmplot(model; plotkwargs...)
save(plotsdir("daisy_step005.png"), plt2)
println("✅ График сохранён: ", plotsdir("daisy_step005.png"))
```

### Шаг 3: После 45 итераций

```julia
step!(model, 40)
println("\nШаг 45: После 45 итераций")
plt3, _ = abmplot(model; plotkwargs...)
save(plotsdir("daisy_step045.png"), plt3)
println("✅ График сохранён: ", plotsdir("daisy_step045.png"))

println("\n🎉 Базовая визуализация завершена!")
```
